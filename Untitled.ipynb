```python
import pandas as pd
from google_play_scraper import Sort, reviews
```

Untuk memperbaiki error `NameError: name 'df_reviews' is not defined`, kita perlu mengambil data ulasan terlebih dahulu. Saya akan menggunakan contoh ID aplikasi 'com.whatsapp' untuk demonstrasi.

Anda dapat mengganti `'com.whatsapp'` dengan ID aplikasi yang Anda inginkan.

```python
app_id = 'id.go.pajak.djp' # Ganti dengan ID aplikasi yang ingin Anda scrape

print(f"Mencoba mengambil ulasan untuk aplikasi '{app_id}'...")

result, continuation_token = reviews(
    app_id,
    lang='id',
    country='id',
    sort=Sort.NEWEST,
    count=1000
)

df_reviews = pd.DataFrame(result)

if df_reviews.empty:
    print(f"Tidak ada ulasan yang ditemukan untuk aplikasi '{app_id}' dengan pengaturan saat ini.")
    # Initialize an empty DataFrame with expected columns to prevent KeyError/NameError in downstream cells
    df_reviews = pd.DataFrame(columns=['userName', 'content', 'score', 'sentiment'])
    print("DataFrame df_reviews diinisialisasi sebagai kosong dengan kolom placeholder.")
elif 'score' not in df_reviews.columns:
    print(f"Kolom 'score' tidak ditemukan dalam ulasan yang diambil untuk aplikasi {app_id}.")
    df_reviews['sentiment'] = 'unknown' # Assign a default if score is missing
    print(f"Berhasil mengambil {len(df_reviews)} ulasan untuk aplikasi {app_id}, namun kolom 'score' tidak tersedia untuk analisis sentimen. Kolom 'sentiment' diisi dengan 'unknown'.")
    display(df_reviews.head())
else:
    df_reviews['sentiment'] = df_reviews['score'].apply(lambda x: 'positive' if x >= 4 else ('neutral' if x == 3 else 'negative'))
    print(f"Berhasil mengambil {len(df_reviews)} ulasan untuk aplikasi {app_id}.")
    display(df_reviews.head())
```

```python
print('Distribusi Sentimen:')
if not df_reviews.empty and 'sentiment' in df_reviews.columns:
    print(df_reviews['sentiment'].value_counts())
else:
    print("Tidak ada data ulasan atau kolom 'sentiment' untuk analisis distribusi sentimen.")

print('\nContoh Ulasan dengan Sentimen:')

if not df_reviews.empty:
    display_cols_candidate = ['userName', 'content', 'score', 'sentiment']
    existing_display_cols = [col for col in display_cols_candidate if col in df_reviews.columns]

    if existing_display_cols:
        display(df_reviews[existing_display_cols].head(10))
    else:
        print("Tidak ada kolom yang relevan (userName, content, score, sentiment) untuk ditampilkan.")
else:
    print("Tidak ada ulasan untuk ditampilkan.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for the plot
sns.set_style('whitegrid')

# Create a bar plot of sentiment distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='sentiment', data=df_reviews, palette='viridis', order=['negative', 'neutral', 'positive']) # Changed order to match actual sentiment labels
plt.title('Distribusi Sentimen Ulasan Google Play', fontsize=16)
plt.xlabel('Sentimen', fontsize=12)
plt.ylabel('Jumlah Ulasan', fontsize=12)
plt.show()
```
